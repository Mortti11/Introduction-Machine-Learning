```python
# library for GUI
import tkinter as tk
from tkinter import messagebox  

import numpy as n
# For loading the scalers
import joblib  
```

```python
# loading the model
model = joblib.load('energy_consumption_model.pkl')
# Loading the scalers
scaler = joblib.load('scaler.pkl')
```

```python
# Using Tkinter for GUI 
# Creating the function for getting the values from the user
def get_values():
    try:
        # getting the values from the user
        Press_mm_hg = float(press_mm_hg_entry.get())
        Laundry_Humidity = float(laundry_humidity_entry.get())
        Teenager_Room_Humidity = float(teenager_room_humidity_entry.get())
        Parents_Room_Humidity = float(parents_room_humidity_entry.get())
        Laundry_Temperature = float(laundry_temperature_entry.get())
        Office_Temperature = float(office_temperature_entry.get())
        Teenager_Room_Temperature = float(teenager_room_temperature_entry.get())
        Outside_Temperature = float(outside_temperature_entry.get())
        Dew_Point_Temperature = float(dew_point_temperature_entry.get())
        hour = int(hour_entry.get())
        
        # Combing the values into a single array
        values = n.array([[Press_mm_hg, Laundry_Humidity, Teenager_Room_Humidity, Parents_Room_Humidity, 
                           Laundry_Temperature, Office_Temperature, Teenager_Room_Temperature, 
                           Outside_Temperature, Dew_Point_Temperature, hour]])
        
        # Scale the values
        scaled_values = scaler.transform(values)
        
        # Predict the energy consumption
        prediction = model.predict(scaled_values)
        
        # Display the prediction
        messagebox.showinfo("Prediction", f"Predicted Energy Consumption: {prediction[0]}")
    except Exception as e:
        messagebox.showerror("Error", str(e))

# Creating the main window
window = tk.Tk()
window.title("Energy Consumption Prediction")
window.geometry("500x500")

# Creating the labels
press_mm_hg_label = tk.Label(window, text="Press_mm_hg")
press_mm_hg_label.pack()
press_mm_hg_entry = tk.Entry(window)
press_mm_hg_entry.pack()

laundry_humidity_label = tk.Label(window, text="Laundry_Humidity")
laundry_humidity_label.pack()
laundry_humidity_entry = tk.Entry(window)
laundry_humidity_entry.pack()

teenager_room_humidity_label = tk.Label(window, text="Teenager_Room_Humidity")
teenager_room_humidity_label.pack()
teenager_room_humidity_entry = tk.Entry(window)
teenager_room_humidity_entry.pack()

parents_room_humidity_label = tk.Label(window, text="Parents_Room_Humidity")
parents_room_humidity_label.pack()
parents_room_humidity_entry = tk.Entry(window)
parents_room_humidity_entry.pack()

laundry_temperature_label = tk.Label(window, text="Laundry_Temperature")
laundry_temperature_label.pack()
laundry_temperature_entry = tk.Entry(window)
laundry_temperature_entry.pack()

office_temperature_label = tk.Label(window, text="Office_Temperature")
office_temperature_label.pack()
office_temperature_entry = tk.Entry(window)
office_temperature_entry.pack()

teenager_room_temperature_label = tk.Label(window, text="Teenager_Room_Temperature")
teenager_room_temperature_label.pack()
teenager_room_temperature_entry = tk.Entry(window)
teenager_room_temperature_entry.pack()

outside_temperature_label = tk.Label(window, text="Outside_Temperature")
outside_temperature_label.pack()
outside_temperature_entry = tk.Entry(window)
outside_temperature_entry.pack()

dew_point_temperature_label = tk.Label(window, text="Dew_Point_Temperature")
dew_point_temperature_label.pack()
dew_point_temperature_entry = tk.Entry(window)
dew_point_temperature_entry.pack()

hour_label = tk.Label(window, text="Hour")
hour_label.pack()
hour_entry = tk.Entry(window)
hour_entry.pack()

# Creating the button
predict_button = tk.Button(window, text="Predict", command=get_values)
predict_button.pack()

# Running the main loop
window.mainloop()
```
