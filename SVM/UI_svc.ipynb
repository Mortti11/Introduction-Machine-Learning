```python
import tkinter as tk
from tkinter import messagebox  
import numpy as n
import joblib 
```

```python
# loading the model
model = joblib.load('datasets/best_svc.pkl')
```

```python
# Function to get values and make predictions
def get_default_prediction():
    try:
        # Getting user input from the entries
        LIMIT_BAL = float(limit_bal_entry.get())
        SEX = int(sex_entry.get())
        EDUCATION = int(education_entry.get())
        MARRIAGE = int(marriage_entry.get())
        AGE = int(age_entry.get())
        PAY_SEP = int(pay_sep_entry.get())
        PAY_AUG = int(pay_aug_entry.get())
        PAY_JUL = int(pay_jul_entry.get())
        PAY_JUN = int(pay_jun_entry.get())
        PAY_MAY = int(pay_may_entry.get())
        PAY_APR = int(pay_apr_entry.get())
        PAY_AMT_SEP = float(pay_amt_sep_entry.get())
        PAY_AMT_AUG = float(pay_amt_aug_entry.get())
        PAY_AMT_JUL = float(pay_amt_jul_entry.get())
        PAY_AMT_JUN = float(pay_amt_jun_entry.get())
        PAY_AMT_MAY = float(pay_amt_may_entry.get())
        PAY_AMT_APR = float(pay_amt_apr_entry.get())
        BILL_AMT = float(bill_amt_entry.get())

        # Combine the input values into an array
        input_values = np.array([[LIMIT_BAL, SEX, EDUCATION, MARRIAGE, AGE, PAY_SEP, PAY_AUG,
                                  PAY_JUL, PAY_JUN, PAY_MAY, PAY_APR, PAY_AMT_SEP, PAY_AMT_AUG,
                                  PAY_AMT_JUL, PAY_AMT_JUN, PAY_AMT_MAY, PAY_AMT_APR, BILL_AMT]])

        # Scale the values using the scaler
        

        # Predict default or not
        prediction = model.predict(input_values)
        labels = ['Did not default (No)', 'Defaulted (Yes)']

        # Show the result in a message box
        messagebox.showinfo("Prediction", f"The customer {labels[int(prediction[0])]}")

    except Exception as e:
        # Handle errors
        messagebox.showerror("Error", str(e))

# Creating the main Tkinter window
window = tk.Tk()
window.title("Default Payment Prediction")
window.geometry("500x800")

# Creating input fields
fields = [
    ('LIMIT_BAL', 'Credit Limit'),
    ('SEX', 'Gender (1=Male, 2=Female)'),
    ('EDUCATION', 'Education (1=Graduate, 2=University, 3=High School, 4=Others)'),
    ('MARRIAGE', 'Marital Status (1=Married, 2=Single, 3=Others)'),
    ('AGE', 'Age'),
    ('PAY_SEP', 'Repayment Status (Sep)'),
    ('PAY_AUG', 'Repayment Status (Aug)'),
    ('PAY_JUL', 'Repayment Status (Jul)'),
    ('PAY_JUN', 'Repayment Status (Jun)'),
    ('PAY_MAY', 'Repayment Status (May)'),
    ('PAY_APR', 'Repayment Status (Apr)'),
    ('PAY_AMT_SEP', 'Payment Amount (Sep)'),
    ('PAY_AMT_AUG', 'Payment Amount (Aug)'),
    ('PAY_AMT_JUL', 'Payment Amount (Jul)'),
    ('PAY_AMT_JUN', 'Payment Amount (Jun)'),
    ('PAY_AMT_MAY', 'Payment Amount (May)'),
    ('PAY_AMT_APR', 'Payment Amount (Apr)'),
    ('BILL_AMT', 'Bill Amount'),
]

entries = {}

for field, label in fields:
    lbl = tk.Label(window, text=label)
    lbl.pack()
    entry = tk.Entry(window)
    entry.pack()
    entries[field] = entry

limit_bal_entry = entries['LIMIT_BAL']
sex_entry = entries['SEX']
education_entry = entries['EDUCATION']
marriage_entry = entries['MARRIAGE']
age_entry = entries['AGE']
pay_sep_entry = entries['PAY_SEP']
pay_aug_entry = entries['PAY_AUG']
pay_jul_entry = entries['PAY_JUL']
pay_jun_entry = entries['PAY_JUN']
pay_may_entry = entries['PAY_MAY']
pay_apr_entry = entries['PAY_APR']
pay_amt_sep_entry = entries['PAY_AMT_SEP']
pay_amt_aug_entry = entries['PAY_AMT_AUG']
pay_amt_jul_entry = entries['PAY_AMT_JUL']
pay_amt_jun_entry = entries['PAY_AMT_JUN']
pay_amt_may_entry = entries['PAY_AMT_MAY']
pay_amt_apr_entry = entries['PAY_AMT_APR']
bill_amt_entry = entries['BILL_AMT']

# Create a Predict button
predict_button = tk.Button(window, text="Predict", command=get_default_prediction)
predict_button.pack()

# Run the main loop
window.mainloop()
```
